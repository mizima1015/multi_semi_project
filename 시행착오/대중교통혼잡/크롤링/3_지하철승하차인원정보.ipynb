## 데이터 불러오기
### 샘플 데이터 확인
- 서울교통공사의 각 호선 구간에 대한 정보만 나와있음.
- 1호선: 청량리~서울역
- 매월 5일 업데이트된다고 함

```python
## https://data.seoul.go.kr/dataList/OA-12252/S/1/datasetView.do 서울시 지하철 호선별 역별 시간대별 승하차 인원 정보
import requests

url= "http://openapi.seoul.go.kr:8088/69576567536c736f3133357542784147/json/CardSubwayTime/1/1000/202307/"
req = requests.get(url)
print(req)
```

```python
json_df = req.json()
# json_df
```

```python
import pandas as pd

sub = json_df['CardSubwayTime']['row']
subway_1 = pd.DataFrame(sub)
print(subway_1.head(1))
```

### 2023년 07월 데이터 불러오기

```python
subway_1
```

- 1호선~9호선만 남기고 삭제

```python
subway_1 = subway_1[subway_1['LINE_NUM'].str.contains('호선')]
subway_1
```

```python
subway_1 = subway_1[~subway_1['LINE_NUM'].str.contains('공항')]
subway_1
```

- WORK_DT(작업일자) 컬럼삭제

```python
subway_1 = subway_1.drop(['WORK_DT'], axis = 1)
subway_1 
```

- 컬럼명 바꾸기(우리말로 하면 너무 길어지니까 조금 나중에 해볼까..)
