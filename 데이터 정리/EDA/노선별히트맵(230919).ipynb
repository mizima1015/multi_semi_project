## 데이터 불러오기
- 승준님이 작업한 노선별 최종 혼잡도

```python
import pandas as pd
i = 1

R_CSV_F =  f"/content/drive/My Drive/[멀티캠]데이터분석&엔지니어취업캠프/data/세미프로젝트_최종혼잡도/은채님{i}호선.csv"
R_CSV = pd.read_csv(R_CSV_F)
R_CSV
```

## 2022년만 추출

```python
R_CSV = R_CSV[R_CSV['YEAR'] == 2022]
R_CSV
```

## 표준편차로 데이터 합치기

```python
cols = R_CSV.columns[6:]

result = R_CSV.groupby(['STATION', 'LINE'])[cols].std().reset_index()

result.sort_values(by='LINE').tail(10)
```

## 그래프

```python
import seaborn as sns
import matplotlib.pyplot as plt

# LINE 컬럼 삭제
result.drop('LINE', axis=1, inplace=True)

# 데이터를 'STATION', 'TIME_00', 'CONGESTION' 열 형식으로 변환
melted_data = result.melt(id_vars="STATION", var_name="TIME_00", value_name="CONGESTION")

plt.figure(figsize=(17, 10))
pivot_table = melted_data.pivot_table(index='STATION', columns='TIME_00', values='CONGESTION')
sns.heatmap(pivot_table, annot=False, cmap='YlGnBu', linewidths=.5,
            xticklabels=True,
            yticklabels=True,
            cbar=True,
            fmt=".1f",
            annot_kws={"size": 10},
            cbar_kws={"label": "Congestion Level"},
            square=True)

plt.xticks(rotation=45, horizontalalignment='right')
plt.yticks(rotation=0, horizontalalignment='right')
plt.title(f'2022년 {i}호선 출발역별 시간대별 혼잡도 히트맵')
plt.show()
```

```python
import matplotlib.pyplot as plt

# 'CONGESTION' 열을 숫자로 변환
melted_data['CONGESTION'] = melted_data['CONGESTION'].astype(float)

# 데이터를 리스트 형식으로 변환
data_array = melted_data.pivot(index='STATION', columns='TIME_00', values='CONGESTION').values

# 역 이름과 시간대 레이블 설정
station_labels = melted_data['STATION'].unique()
time_labels = melted_data['TIME_00'].unique()

# 히트맵 그리기
plt.figure(figsize=(10, 6))
plt.imshow(data_array, cmap='YlGnBu', aspect='auto')
plt.colorbar(label='Congestion Level')

# x축과 y축 레이블 설정
plt.xticks(range(len(time_labels)), time_labels, rotation=45, horizontalalignment='right')
plt.yticks(range(len(station_labels)), station_labels, rotation=0, horizontalalignment='right')

plt.xlabel('TIME_00')
plt.ylabel('STATION')
plt.title('Congestion Heatmap')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 'CONGESTION' 열을 숫자로 변환
melted_data['CONGESTION'] = melted_data['CONGESTION'].astype(float)

# 데이터를 리스트 형식으로 변환
data_list = melted_data.values.tolist()

# 데이터를 행렬로 변환
station_labels = list(set([row[0] for row in data_list]))
time_labels = list(set([row[1] for row in data_list]))
time_labels = ['TIME_05', 'TIME_06', 'TIME_07', 'TIME_08', 'TIME_09', 'TIME_10', 'TIME_11', 'TIME_12', 'TIME_13', 'TIME_14', 'TIME_15', 'TIME_16',
               'TIME_17', 'TIME_18', 'TIME_19', 'TIME_20', 'TIME_21', 'TIME_22', 'TIME_23', 'TIME_00']
congestion_values = np.zeros((len(station_labels), len(time_labels)))

for row in data_list:
    station_idx = station_labels.index(row[0])
    time_idx = time_labels.index(row[1])
    congestion_values[station_idx][time_idx] = row[2]

# 히트맵 그리기
plt.figure(figsize=(10, 6))
plt.imshow(congestion_values, cmap='YlGnBu', aspect='auto')
plt.colorbar(label='Congestion Level')

# x축과 y축 레이블 설정
plt.xticks(range(len(time_labels)), time_labels, rotation=45, horizontalalignment='right')
plt.yticks(range(len(station_labels)), station_labels, rotation=0, horizontalalignment='right')

plt.xlabel('TIME_00')
plt.ylabel('STATION')
plt.title('Congestion Heatmap')
plt.show()
```
