## 기온 데이터 날짜별 정리

```python
temp.drop(temp.index[0:6], inplace=True)
temp.drop(columns = ['Unnamed: 1'], inplace=True)
temp.drop([244], axis=0, inplace=True)
temp
```

```python
temp_1 = temp.copy()

# 컬럼명 바꾸기 
temp_2 = temp_1.rename(columns=temp_1.iloc[0])
temp_2
```

```python
# 인덱스6 행 삭제
temp_2.drop([6], axis=0, inplace=True)
temp_2
```

```python
# 날짜 앞 \t 삭제
temp_2['날짜'] = temp_2['날짜'].str.replace('\t', ' ') 
temp_2
```

### 결측치 채우기
- 기상청 홈페이지에서 보니 결측치의 강수량은 0

```python
temp_rain_data = temp_2.fillna(0)
temp_rain_data
```

```python
# 인덱스 초기화 
temp_rain_data.reset_index(inplace = True)
temp_rain_data
```

## 기상조건 표시
맑음 / 약한눈비 / 강한눈비(110mm이상, 신적설 5cm이상) / 한파(최저기온 -12이하), 폭염(최고온도 33도 이상)

```python
print(temp_rain_data.dtypes)
```

```python
temp_rain_data['평균기온(℃)'] = pd.to_numeric(temp_rain_data['평균기온(℃)'], errors='coerce')
temp_rain_data['최저기온(℃)'] = pd.to_numeric(temp_rain_data['최저기온(℃)'], errors='coerce')
temp_rain_data['최고기온(℃)'] = pd.to_numeric(temp_rain_data['최고기온(℃)'], errors='coerce')
temp_rain_data['강수량(mm)'] = pd.to_numeric(temp_rain_data['강수량(mm)'], errors='coerce')
print(temp_rain_data.dtypes)
```

```python
new_values = []
for i in range(len(temp_rain_data)):
    if temp_rain_data['최고기온(℃)'][i] >= 33 or temp_rain_data['최저기온(℃)'][i] <= -12:
        new_values.append("폭염/한파")
    elif temp_rain_data['강수량(mm)'][i] >= 110:
        new_values.append("강한눈비")
    elif temp_rain_data['최저기온(℃)'][i] < -12:
        new_values.append("폭염/한파")
    elif temp_rain_data['강수량(mm)'][i] > 0 and temp_rain_data['강수량(mm)'][i] < 110:
        new_values.append("약한눈비")
    else:
        new_values.append("맑음")
temp_rain_data['기상조건'] = new_values

print(temp_rain_data)
```

## 날짜 형식 바꾸기

```python
temp_rain_data['날짜']
```

```python
import datetime
temp_rain_data['날짜'] = pd.to_datetime(temp_rain_data['날짜'])
temp_rain_data['날짜'].head(1)
```

- 연/월/일 컬럼 만들기

```python
temp_rain_data['연'] = temp_rain_data['날짜'].dt.year
temp_rain_data['월'] = temp_rain_data['날짜'].dt.month
temp_rain_data['일'] = temp_rain_data['날짜'].dt.day
temp_rain_data.head(1)
```

- 요일 추가

```python
temp_rain_data['요일'] = temp_rain_data['날짜'].dt.weekday
temp_rain_data
```

```python
# 0~6번 인덱스 위치에 각 한글 요일 이름이 담긴 리스트
weekday_list = ['월', '화', '수', '목', '금', '토', '일']

# apply 함수로 숫자 -> 요일 이름으로 변환
temp_rain_data['요일'] = temp_rain_data.apply(lambda x : weekday_list[x['요일']], axis = 1)

temp_rain_data
```

```python
# 날짜 컬럼 삭제
temp_rain_df = temp_rain_data.drop(['날짜'], axis=1).copy()
temp_rain_df
```

```python
# 컬럼 순서 바꾸기
temp_rain_df = temp_rain_df[['연', '월', '일', '요일', '기상조건', '평균기온(℃)', '최저기온(℃)', '최고기온(℃)', '강수량(mm)']]
temp_rain_df
```

## 저장하기

```python
temp_rain_df.to_csv("[기상_102]_날짜_요일_기상조건_2023.csv", index=False)
```
