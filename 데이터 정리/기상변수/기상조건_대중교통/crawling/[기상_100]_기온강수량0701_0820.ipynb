## 데이터 불러오기

```python
import pandas as pd
temp = pd.read_csv("서울기온_2207_2308.csv")
rain = pd.read_csv("서울_강수량_2207_2308.csv")
```

```python
print(temp.info())
print(rain.info())
```

## 기온 데이터 날짜별 정리

```python
temp.drop(temp.index[0:6], inplace=True)
# temp.drop(columns = ['Unnamed: 1'], inplace=True)
temp.drop([423], axis=0, inplace=True)
temp
```

```python
temp_1 = temp.copy()

# 컬럼명 바꾸기 
temp_2 = temp_1.rename(columns=temp_1.iloc[0])
temp_2
```

```python
# 인덱스6 행 삭제
temp_2.drop([6], axis=0, inplace=True)
temp_2
```

```python
# 날짜 앞 \t 삭제
temp_2['날짜'] = temp_2['날짜'].str.replace('\t', ' ') 
temp_2
```

```python
# 날짜 컬럼을 인덱스로 두기
temp_3 = temp_2.set_index('날짜')
temp_3.drop(columns = ['지점'], inplace=True)
temp_3
```

## 다시!!!!!! 데이터 불러오기

```python
df_seoul = pd.read_csv("서울기온강수량_2207_2308.csv")
df_seoul
```

```python
df_seoul.info()
```

```python
df_seoul.drop(df_seoul.index[0:6], inplace=True)
df_seoul.drop(columns = ['Unnamed: 1'], inplace=True)
df_seoul.drop([423], axis=0, inplace=True)
df_seoul
```

```python
# 컬럼명 바꾸기
df_seoul_1 = df_seoul.rename(columns=df_seoul.iloc[0])
df_seoul_1.drop([6], axis=0, inplace=True)
df_seoul_2 = df_seoul_1.copy()
df_seoul_2
```

```python
# 날짜 컬럼에서 \t 지우기
df_seoul_2['날짜'] = df_seoul_2['날짜'].str.replace('\t', ' ') 
df_seoul_3 = df_seoul_2.copy()
df_seoul_3
```

```python
# 날짜 컬럼을 인덱스로 두기
df_seoul_4 = df_seoul_3.set_index('날짜')
df_seoul_4
```

```python
# 강수량 컬럼의 결측치 0으로 두기 
df_seoul_5 = df_seoul_4.fillna(0)
df_seoul_5
```
