#### 시간 쪼개기

- 우선 열 이름 구성을 보자
    - 9:30~ (%)
    - 이런 형태인데, 뒤의 30을 기준으로 앞이 같은 녀석들을 합치고, 그 후 시간대 나눠합쳐도 좋을 것 같다.

```python
# 30분 단위 묶기

time_columns = Subw_d01.columns[:-5]  # 시간대 열들 선택 (마지막 5개 열은 다른 정보)
for i in range(0, len(time_columns), 2):
    if i + 1 < len(time_columns):  # 인덱스 범위 확인
        start_col = time_columns[i]
        end_col = time_columns[i + 1]
        new_col_name = start_col.split('~')[0]  # 시간대의 시작만 남김
        Subw_d01[new_col_name] = (pd.to_numeric(Subw_d01[start_col]) + pd.to_numeric(Subw_d01[end_col])) / 2  # 숫자로 변환 후 평균 계산
Subw_d01 = Subw_d01.drop(time_columns, axis=1)  # 원래 열 제거

print(Subw_d01)
```
