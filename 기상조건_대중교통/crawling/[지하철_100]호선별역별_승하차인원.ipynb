```python
import requests
import json
import pandas as pd

SERVICE_KEY = '69576567536c736f3133357542784147'

date = pd.date_range(start="20210101", end="20211231")
date_21 = date.strftime("%Y%m%d")

data = None
for j in range(len(date_21)):
    url = f'http://openapi.seoul.go.kr:8088/{SERVICE_KEY}/Json/CardSubwayStatsNew/1/1000/{date_21[j]}'
    # print(url)
    req = requests.get(url)
    content = req.json()
    con = content['CardSubwayStatsNew']['row']
    result = pd.DataFrame(con)
    data = pd.concat([data, result])
data = data.reset_index(drop=True)
data.info()
```

```python
data
```

- 1호선~9호선만 남기고 삭제

```python
data = data[data['LINE_NUM'].str.contains('호선')]
data
```

```python
data = data[~data['LINE_NUM'].str.contains('공항')]
data
```

- WORK_DT(작업일자) 컬럼삭제

```python
data = data.drop(['WORK_DT'], axis = 1)
data 
```

### 컬럼명 바꾸기
- USE_DT : 사용일자 / LINE_NUM : 호선명 / SUB_STA_NM : 지하철역 / RIDE_PASGR_NUM :승차총승객수 / 	ALIGHT_PASGR_NUM : 	하차총승객수

```python
data.columns = ['사용일자', '호선명', '지하철역', '승차총승객수', '하차총승객수']
data
```

### 날짜 바꾸기 
- 연/월/일 컬럼 만들기

```python
data_1 = data.copy()
data_1.info()
```

```python
import datetime
data['사용일자'] = pd.to_datetime(data['사용일자'])
data['사용일자'].head(1)
```

```python
data['연'] = data['사용일자'].dt.year
data['월'] = data['사용일자'].dt.month
data['일'] = data['사용일자'].dt.day
data.head(1)
```

- 요일 추가

```python
data['요일'] = data['사용일자'].dt.weekday
data
```

```python
# 0~6번 인덱스 위치에 각 한글 요일 이름이 담긴 리스트
weekday_list = ['월', '화', '수', '목', '금', '토', '일']

# apply 함수로 숫자 -> 요일 이름으로 변환
data['요일'] = data.apply(lambda x : weekday_list[x['요일']], axis = 1)

data
```

```python
# 컬럼 순서 바꾸기
data = data[['사용일자', '연', '월', '일', '요일', '호선명', '지하철역', '승차총승객수', '하차총승객수']]
data
```

```python
data.to_csv("[지하철_100]호선별역별_승하차정보_2021.csv", index=False)
```
