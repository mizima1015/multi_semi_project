## 데이터 불러오기
### 샘플 데이터 확인

```python
## https://data.seoul.go.kr/dataList/10283/S/2/datasetView.do 서울시 통근·통학시 이용하는 교통수단 통계(2022년)

import pandas as pd
df = pd.read_csv("[지하철_101]통근통학_교통수단.csv", encoding='utf-8')
df.head(1)
```

### 컬럼 확인
- 구분별(1) : 서울시(모두 동일)
- 구분별(2) : 소계, 성별, 연령별, 학력별, 소득별, 자치구별
- 구분별(3) : 구분별(2)의 세부항목
- 나머지는 2022년 동일한 컬럼명

```python
df.columns
```

- 구분별(1) 컬럼 삭제

```python
df = df.drop(['구분별(1)'], axis=1)
df.head(1)
```

- 인덱스0을 컬럼명으로 지정

```python
df_1 = df.rename(columns=df.iloc[0])
df_1.head(1)
```

- 인덱스 0 행 삭제

```python
# df_1.drop([0], axis=0, inplace=True)
df_1.head(5)
```

```python
df_1.info()
```

- 컬럼명 수정

```python
df_1.rename(columns={'구분별(2)' : '대분류'}, inplace=True)
df_1.rename(columns={'구분별(3)' : '분류'}, inplace=True)
df_1.head(10)
```

### 지역구 통계
- 지역구 데이터 추출

```python
df_z = df_1[df_1['대분류'] == '지역소분류'].copy()
df_z
```

- '대분류' 컬럼 삭제

```python
# df_z.drop(columns = '대분류', inplace=True)
df_z.head(1)
```

- '-'를 0으로 바꾸기

```python
df_z = df_z.replace('-', 0)
```

- '분류' 열을 인덱스로 지정

```python
# df_z.set_index('분류', inplace=True)
df_zz = df_z.T
df_zz.columns
```
