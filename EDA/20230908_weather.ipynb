```python
import pandas as pd
import matplotlib.pyplot as plt

# 'data' 데이터프레임에서 '날짜' 열을 문자열로 변환
data['날짜'] = data['날짜'].astype(str)

# "일시" 값을 포함한 행을 필터링하고 제거합니다.
data = data[~data['날짜'].str.contains('일시')]

# '날짜' 열을 datetime 형식으로 변환
data['날짜'] = pd.to_datetime(data['날짜'], format='%Y-%m-%d')

# 데이터 프레임에서 필요한 열만 선택
rain_data = data[['날짜', '강수량']]

# 'merged_data' 데이터프레임에서 '날짜' 열을 문자열로 변환
merged_data['날짜'] = merged_data['날짜'].astype(str)

# "일시" 값을 포함한 행을 필터링하고 제거합니다.
merged_data = merged_data[~merged_data['날짜'].str.contains('일시')]

# '날짜' 열을 datetime 형식으로 변환
merged_data['날짜'] = pd.to_datetime(merged_data['날짜'], format='%Y-%m-%d')

# 데이터 프레임에서 필요한 열만 선택
passenger_data = merged_data[['날짜', '합 계']]

# 원하는 기간 선택 (2023-01 ~ 2023-07)
start_date = '2023-01'
end_date = '2023-07'
rain_data = rain_data[(rain_data['날짜'] >= start_date) & (rain_data['날짜'] <= end_date)]
passenger_data = passenger_data[(passenger_data['날짜'] >= start_date) & (passenger_data['날짜'] <= end_date)]

# '날짜' 열에서 년도와 월을 추출하여 새로운 열 추가
rain_data.loc[:, '년도'] = rain_data['날짜'].dt.year
rain_data.loc[:, '월'] = rain_data['날짜'].dt.month
passenger_data.loc[:, '년도'] = passenger_data['날짜'].dt.year
passenger_data.loc[:, '월'] = passenger_data['날짜'].dt.month

# 각 달의 데이터 추출
monthly_rainfall = rain_data.groupby(['년도', '월'])['강수량'].sum().reset_index()
monthly_passengers = passenger_data.groupby(['년도', '월'])['합 계'].sum().reset_index()

# 파이차트 그리기
fig, axs = plt.subplots(1, 2, figsize=(12, 6))

# 'data' 데이터프레임 파이차트
labels = [f'{year}-{month:02d}' for year, month in zip(monthly_rainfall['년도'], monthly_rainfall['월'])]
sizes = monthly_rainfall['강수량']
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6', '#c2f0c2']
axs[0].pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
axs[0].axis('equal')
axs[0].set_title('날짜별 강수량 비교')

# 'merged_data' 데이터프레임 파이차트
labels = [f'{year}-{month:02d}' for year, month in zip(monthly_passengers['년도'], monthly_passengers['월'])]
sizes = monthly_passengers['합 계']
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0', '#ffb3e6', '#c2f0c2']

axs[1].pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
axs[1].axis('equal')
axs[1].set_title('강수량에 따른 날짜별 지하철 이용객 변화')  # 타이틀 수정

plt.tight_layout()
plt.show()
```
