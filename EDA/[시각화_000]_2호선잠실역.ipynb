### 데이터 불러오기

```python
import pandas as pd

wet_23 = pd.read_csv("[기상_102]_날짜_요일_기상조건_2023.csv")
sub_23 = pd.read_csv("[지하철_100]호선별역별_승하차정보_2023.csv")
```

### 2호선 잠실역 컬럼 추출

```python
sub_23_js = sub_23[(sub_23['지하철역'] == '잠실(송파구청)') & (sub_23['호선명'] == '2호선')].reset_index()
sub_23_js
```

### 기상조건 추가
- wet_23에서 기상조건 컬럼 추출

```python
wet_23_con = wet_23['기상조건']
wet_23_con 
```

- 데이터프레임 합치기

```python
jsil_2 = pd.concat([sub_23_js, wet_23_con], axis=1)
jsil_2.drop(columns='index', axis=1)
jsil_2.dropna(inplace=True)
jsil_2
```

# EDA

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
```

- 한글 폰트 설정

```python
import platform
print(platform.system())
# 운영체제별 한글 폰트 설정
if platform.system() == 'Darwin': # Mac 환경 폰트 설정
    plt.rc('font', family='AppleGothic')
elif platform.system() == 'Windows': # Windows 환경 폰트 설정
    plt.rc('font', family='Malgun Gothic')

plt.rc('axes', unicode_minus=False) # 마이너스 폰트 설정


# 글씨 선명하게 출력하는 설정
%config InlineBackend.figure_format = 'retina'
```

## 2호선 잠실역 관련

- 전체 승차총승객수 분포

```python
sns.displot(jsil_2['승차총승객수'],
           color = "#FC7F77")

plt.show()
```

- 기상조건별 2호선 잠실역 승차총승객수

```python
jsil_2.groupby(['기상조건'])['승차총승객수'].mean()
```

```python
plt.figure(figsize = (12, 6))
sns.barplot(x = '기상조건',
            y = '승차총승객수',
            data = jsil_2)
plt.show()
```

- 월별 2호선 잠실역 승차총승객수

```python
jsil_2.groupby(['월'])['승차총승객수'].mean()
```

```python
jsil_2 = jsil_2.astype({'월' : 'int'})
plt.figure(figsize = (12, 6))
sns.barplot(x = '월',
            y = '승차총승객수',
            data = jsil_2, 
            palette="Set2")
plt.show()
```

- 기상조건별 빈도수

```python
plt.figure(figsize = (9, 6))
plt.xticks(rotation = 10)
sns.countplot(x = '기상조건',
              data = jsil_2, 
            palette="Set2")
plt.show()
```

- 요일별 승차총승객수

```python
jsil_2.groupby(['요일'])['승차총승객수'].mean()
```

```python
jsil_2 = jsil_2.astype({'월' : 'int'})
plt.figure(figsize = (12, 6))
sns.barplot(x = '요일',
            y = '승차총승객수',
            data = jsil_2, 
            palette="Set2"
           )
plt.show()
```
